### Алгоритм матрицы взаимной встречаемости пикселя

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import graycomatrix, graycoprops
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.util import img_as_ubyte

def calculate_entropy(glcm):
    glcm_normalized = glcm / np.sum(glcm)
    glcm_nonzero = glcm_normalized[glcm_normalized > 0]
    return -np.sum(glcm_nonzero * np.log2(glcm_nonzero))

def glcm_analysis(image_path, distances=[1], angles=[0]):
    # Загружаем изображение и переводим его в градации серого
    image = imread(image_path)
    if len(image.shape) == 3:  # Если изображение цветное
        image = rgb2gray(image)
    image = img_as_ubyte(image)  # Приведение к формату 8-битных значений
    
    glcm = graycomatrix(image, distances=distances, angles=angles, levels=256, symmetric=True, normed=False)
    
    contrast = graycoprops(glcm, 'contrast')[0, 0]
    energy = graycoprops(glcm, 'energy')[0, 0]
    homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]
    entropy = calculate_entropy(glcm[:, :, 0, 0])
    
    features = {
        'contrast': contrast, #мера различий в интенсивности
        'energy': energy, #мера текстурной однородности
        'homogeneity': homogeneity, #равномерность распределения
        'entropy': entropy #мера хаотичности
    }
    
    return glcm[:, :, 0, 0], features

def visualize_glcm(glcm):
    plt.figure(figsize=(8, 6))
    plt.imshow(glcm, cmap='gray', interpolation='nearest')
    plt.title("Матрица взаимной встречаемости пикселя")
    plt.xlabel("Gray Level 1")
    plt.ylabel("Gray Level 2")
    plt.colorbar(label="Frequency")
    plt.show()

image_path = "sick.jpg" 
glcm_matrix, features = glcm_analysis(image_path)

print("Характеристи матрицы взаимной встречаемости пикселя")
print(features)

print("\nМатрица взаимной встречаемости пикселя")
print(glcm_matrix)

visualize_glcm(glcm_matrix)
```

```python
image_path = "norm.jpg" 
glcm_matrix, features = glcm_analysis(image_path)

print("Характеристи матрицы взаимной встречаемости пикселя")
print(features)

print("\nМатрица взаимной встречаемости пикселя")
print(glcm_matrix)

visualize_glcm(glcm_matrix)
```
